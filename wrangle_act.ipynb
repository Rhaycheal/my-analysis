```python
dog_breed = []
conf_list = []
def image(image_predict_clean):
    if image_predict_clean['p1_dog'] == True:
        dog_breed.append(image_predict_clean['p1'])
        conf_list.append(image_predict_clean['p1_conf'])
    elif image_predict_clean['p2_dog'] == True:
        dog_breed.append(image_predict_clean['p2'])
        conf_list.append(image_predict_clean['p2_conf'])
    elif image_predict_clean['p3_dog'] == True:
        dog_breed.append(image_predict_clean['p3'])
        conf_list.append(image_predict_clean['p3_conf'])
    else:
        dog_breed.append('None')
        conf_list.append('None')
```

```python
image_predict_clean.apply(image, axis=1)

image_predict_clean['dog_breed'] = dog_breed
image_predict_clean['conf_list'] = conf_list
```
